# Train Test Split

```python
train = data[data['Date'] < '20150101']
test = data[data['Date'] > '20141231']
```

# Data Cleaning and Preprocessing of Traning data

```python
data = train.iloc[:,2:27]
list1 = [i for i in range(25)]
new_index = [str(i) for i in list1]
data.columns=new_index
data.head()
```

```python
headlines = []
for i in range(0,len(data)):
  headlines.append(' '.join(str(x) for x in data.iloc[i,0:25]))
```

```python
headlines
```

```python
import re
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
lemmitizer = WordNetLemmatizer()
```

```python
corpus=[]
stopword = stopwords.words('english')
stopword.remove('not')
for i in range(0, len(headlines)):
  review = re.sub('[^a-zA-Z0-9]', ' ', headlines[i])
  review = review.lower()
  review = review.split()
  a = []
  for word in review:
    if word not in stopword:
      a.append(lemmitizer.lemmatize(word))
  corpus.append(' '.join(a))
```

# Data Cleaning and Preprocessing of Test data

```python
data = test.iloc[:,2:27]
list1 = [i for i in range(25)]
new_index = [str(i) for i in list1]
data.columns=new_index
data.head()
```

```python
headlines_test = []
for i in range(0, len(data)):
  headlines_test.append(' '.join(str(x) for x in data.iloc[i,2:27]))
```

```python
headlines_test
```

```python
corpus_test=[]
for i in range(0, len(headlines_test)):
  review = re.sub('[^a-zA-Z0-9]', ' ', headlines_test[i])
  review = review.lower()
  review = review.split()
  a = []
  for word in review:
    if word not in stopword:
      a.append(lemmitizer.lemmatize(word))
  corpus_test.append(' '.join(a))
```

# Data Transformation(BOW) and Model Traning

```python
from sklearn.feature_extraction.text import CountVectorizer
countvector=CountVectorizer(ngram_range=(2,2), binary=True)
```

```python
X = countvector.fit_transform(corpus)
```

```python
X.shape
```

```python
X_test = countvector.transform(corpus_test)
X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import numpy as np
```

```python
randomclassifier = RandomForestClassifier(n_estimators=200, criterion='entropy')
randomclassifier.fit(X, train['Label'])
```

```python
y_pred = randomclassifier.predict(X_test)
score = accuracy_score(test['Label'], y_pred)
print(score)
print(classification_report(test['Label'], y_pred))
conf_matrix = confusion_matrix(np.array(test['Label']), y_pred)
print(conf_matrix)
```

```python
import seaborn as sns
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
```

# Data Transformation(TF-IDF) and Model Traning

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidfvectorizer = TfidfVectorizer(ngram_range=(2,2), binary=True)
```

```python
X_tf = tfidfvectorizer.fit_transform(corpus).toarray()
X_tf.shape
```

```python
X_tf_test = TfidfVectorizer(max_features=10000,binary=True).fit_transform(corpus_test).toarray()
X_tf_test.shape
```

```python
model_tfidf = RandomForestClassifier().fit(X_tf,train['Label'])
```

```python
y_pred_tf = model_tfidf.predict(X_tf_test)
score = accuracy_score(y_pred_tf,test['Label'])
print(score)
print(classification_report(test['Label'],y_pred_tf))
```
